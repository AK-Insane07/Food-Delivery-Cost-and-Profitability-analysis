```python
import pandas as pd
```

```python
fd_data=pd.read_csv('/content/food_orders_new_delhi.csv')
print(fd_data)
```

```python
fd_data.info()
```

```python
fd_data.isnull().sum()
```

```python
from datetime import datetime
fd_data['Order Date and Time']=pd.to_datetime(fd_data['Order Date and Time'])
fd_data['Delivery Date and Time']=pd.to_datetime(fd_data['Delivery Date and Time'])

def extract_discount(discount_str):
    if isinstance(discount_str, str):  # Ensure it's a string
        if 'off' in discount_str:
            return float(discount_str.split(' ')[0])
        elif '%' in discount_str:
            return float(discount_str.replace('%', ' ')[0])
        else:
            return 0.0
    return 0.0


fd_data['Discount Percentage'] = fd_data['Discounts and Offers'].apply(lambda x: extract_discount(x))



fd_data['Discount Amount'] = fd_data.apply(lambda x: (x['Order Value'] * x['Discount Percentage'] / 100)
                                                   if x['Discount Percentage'] > 1
                                                   else x['Discount Percentage'], axis=1)


fd_data['Discount Amount'] = fd_data.apply(lambda x: x['Discount Amount'] if x['Discount Percentage'] <= 1
                                                   else x['Order Value'] * x['Discount Percentage'] / 100, axis=1)

print(fd_data[['Order Value', 'Discounts and Offers', 'Discount Percentage', 'Discount Amount']].head(), fd_data.dtypes)
```

```python
fd_data['Total Cost']=fd_data['Delivery Fee'] + fd_data['Payment Processing Fee'] + fd_data['Discount Amount']
fd_data['Revenue']=fd_data['Commission Fee']
fd_data['Profit']=fd_data['Revenue'] - fd_data['Total Cost']

Total_orders=fd_data.shape[0]
Total_revenue=fd_data['Revenue'].sum()
Total_costs=fd_data['Total Cost'].sum()
Total_profit=fd_data['Profit'].sum()



overall_metrics={
    "Total Orders":Total_orders,
    "Total Revenue":Total_revenue,
    "Total Cost":Total_costs,
    "Total Profit":Total_profit
             }

print(overall_metrics)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
plt.hist(fd_data['Profit'], bins=50,color="skyblue",edgecolor='black')
plt.title("Profit distribution per order in food delivery")
plt.xlable('Profit')
plt.ylable('Number of Order')
plt.axvline(fd_data['Profit'].mean(),linestyle='dashed',color='green',linewidth=2 )
plt.show()
```

```python
cost_breakdown=fd_data[['Delivery Fee','Payment Processing Fee','Discount Amount']].sum()
plt.figure(figsize=(7,7))
plt.pie(cost_breakdown,labels=cost_breakdown.index, autopct="%1.1f%%",startangle=140,colors=['gold','red','blue'])
plt.title('Proportion of Total Costs in Food Delivery')
plt.show()
```

```python
totals=['Total Revenue','Total Costs','Total Profit']
values=[Total_revenue,Total_costs,Total_profit]

plt.figure(figsize=(10,6))
plt.bar(totals,values,color=['green','pink','red'])
plt.title('Total Revenue, Costs, and Profit')
plt.ylabel('Amount (INR)')
plt.show()
```

```python
profit_order=fd_data[fd_data['Profit']>0]
profit_order['Commission Percentage']=(profit_order['Commission Fee'] / profit_order['Order Value'] )*100
profit_order['Discount Percentage']=(profit_order['Discount Amount'] / profit_order['Order Value'] )*100

new_avgCP=profit_order['Commission Percentage'].mean()
new_avgDP=profit_order['Discount Percentage'].mean()

print(new_avgCP,new_avgDP)
```

```python

recommended_commission_percentage = 30.0
recommended_discount_percentage = 6.0


fd_data['Simulated Commission Fee'] = fd_data['Order Value'] * (recommended_commission_percentage / 100)
fd_data['Simulated Discount Amount'] = fd_data['Order Value'] * (recommended_discount_percentage / 100)

fd_data['Simulated Total Costs'] = (fd_data['Delivery Fee'] +
                                        fd_data['Payment Processing Fee'] +
                                        fd_data['Simulated Discount Amount'])

fd_data['Simulated Profit'] = (fd_data['Simulated Commission Fee'] -
                                   fd_data['Simulated Total Costs'])


import seaborn as sns

plt.figure(figsize=(14, 7))

sns.kdeplot(fd_data['Profit'], label='Actual Profitability', fill=True, alpha=0.5, linewidth=2)

sns.kdeplot(fd_data['Simulated Profit'], label='Estimated Profitability with Recommended Rates', fill=True, alpha=0.5, linewidth=2)

plt.title('Comparison of Profitability in Food Delivery: Actual vs. Recommended Discounts and Commissions')
plt.xlabel('Profit')
plt.ylabel('Density')
plt.legend(loc='upper left')
plt.show()
```
